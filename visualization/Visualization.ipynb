# Visualization
First, we visualize an arbitrary qubit state on the so-called Bloch sphere.

Then, we also present two types of visualization (using Husimi and Wigner functions) for some typical quantum states.

```python
import numpy as np
from tqix import *
```

# 1. Bloch sphere
For a qubit state $|\psi\rangle$, we can represent it as a complex superposition of two basis states $|0\rangle$ and $|1\rangle$:
$$|\psi\rangle = \cos \frac{\theta}{2}|0\rangle +e^{i\phi}\sin \frac{\theta}{2}|1\rangle, $$
where $\theta$ and $\phi$ are polar and azimuthal angles, respectively.

```python
qubit = random(2)
bloch_sphere(qubit)
```

# 2. Husimi Q function
This is a quasi-probability function of a density matrix $\hat{\rho}$ defined as
$$Q(\alpha)=\frac{1}{\pi}\langle\alpha |\hat{\rho}|\alpha \rangle$$
where $|\alpha \rangle$ are coherent states.

## a. For 2D

```python
state = coherent(20,1+1j)
xrange = [0,3]
yrange = [0,3]
husimi_2d(state, xrange,yrange,fname='coherent_husimi_2d.eps')
```

## b. For 3D

```python
state = coherent(20,1+1j)
xrange = [-2,4]
yrange = [-2,4]
husimi_3d(state, xrange,yrange,fname='coherent_husimi_3d.eps')
```

## c. For 3D Spin

```python
j = 37/2
theta = 0.5*np.pi
phi = -0.3*np.pi
psi = spin_coherent(j, theta, phi)
##################
THETA = [0, np.pi]
PHI = [0, 2* np.pi]
husimi_spin_3d(psi, THETA ,PHI ,cmap = cmindex
(1),fname = 'husimi_spin_coherent.eps')
```

# 3. Wigner function
This is also a quasi-probability function of a quantum state $|\psi\rangle $ which is defined as
$$W(x,p) = \frac{1}{\pi \hbar} \int_{-\infty}^{\infty}\psi ^* (x+y)\psi (x-y) e^{2ipy/\hbar} dy $$
where $x$ and $p$ are position and momentum, respectively.

## a. For 2D

```python
state = coherent(20,1)
xrange = [-1,3]
yrange = [-2,2]
wigner_2d(state, xrange,yrange,fname='coherent_wigner_2d.eps')
```

## b. For 3D

```python
state = coherent(20,1)
xrange = [-1,3]
yrange = [-2,2]
wigner_3d(state, xrange,yrange,fname='coherent_wigner_3d.eps')
```

## c. For 3D Spin

```python
j = 37/2
theta = 0.5*np.pi
phi = -0.3*np.pi
psi = spin_coherent(j, theta, phi)
##################
THETA = [0, np.pi]
PHI = [0, 2* np.pi]
wigner_spin_3d(psi, THETA ,PHI ,cmap = cmindex
(1),fname = 'wigner_spin_coherent.eps')
```
