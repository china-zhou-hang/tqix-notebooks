# Backends - Cumulative Distribution Function and Monte Carlo simulations

```python
from tqix import *
import numpy as np
import matplotlib.pyplot as plt
import time
```

# 1. Initiation of a function
Here we generate the function $f(x)=e^{-x}$ for an example.

```python
def func(n):
    fx = []
    x = np.linspace(0,5,n)
    for i in range(n):
        fx.append(np.exp(-x[i]))
    return fx
```

# 2. Number of samples and iterations

```python
samp = 1000 #number of sample
ite = 1000 #number iteration
```

```python
fx = func(samp)
mc_sim = []
cdf_sim = []
```

# 3. Monte Carlo simulation

```python
for i in fx:
    mc_sim.append(mc(i,ite))
```

# 4. Cumulative distribution function (CDF) simulation

```python
for i in fx:
    temp = []
    for j in range(ite):
        temp.append(randunit()/i)
    cdf_sim.append(cdf(temp))
```

# 5. Showing the results of simulation for $f(x) = e^{-x}$

```python
x = np.linspace(0,5,samp)
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x,mc_sim,'b+')
ax.plot(x,cdf_sim,'g.')
ax.plot(x, fx, 'r')
ax.legend(("mc","cdf","fx"))
ax.set_xlabel('x')
ax.set_ylabel('exponential');
plt.savefig('fig5.eps')
plt.show()
```

# 6. Elapsed time of simulations

```python
mc_times = []
cdf_times = []
ites = []
```

```python
for ite in range(1000,11000,1000):
    mc_start = time.time()
    for i in fx:
        mc_sim.append(mc(i,ite))
    mc_stop = time.time()
    delta = mc_stop - mc_start
    mc_times.append(delta)
    
    cdf_start = time.time()
    for i in fx:
        temp = []
        for j in range(ite):
            temp.append(randunit()/i)
        cdf_sim.append(cdf(temp))
    cdf_stop = time.time()
    delta = cdf_stop - cdf_start
    cdf_times.append(delta)
    ites.append(ite)
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))
ax1.plot(ites, mc_times, 'r')
ax1.plot(ites, cdf_times,'b--')
ax1.legend(("mc","cdf"))
ax1.set_xlabel('iteration')
ax1.set_ylabel('time (s)')
plt.show()
```
